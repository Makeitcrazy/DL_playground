```python
# To apply input tensor and kernel tensor
import torch
from torch.nn.parameter import Parameter 
```

```python
# To initialize a 4 x 4 matrix with parameters
A = torch.tensor([[[[1,2,0,1],
                    [-1,0,3,2],
                    [1,3,0,1],
                    [2,-2,1,0]]]]).float()

B = torch.tensor([[[[3,4,1,1],
                    [-3,3,5,2],
                    [4,1,6,2],
                    [3,-5,1,3]]]]).float()

# torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros', device=None, dtype=None)
# Args:
# in_channels = no. input image; out_channels = no. output image; kernal_size = 2 = 2 x 2 
conv2d = torch.nn.Conv2d(1, 1, kernel_size= 1, bias = False) # What does this function mean? 
# Different weight [][] produces different dimensions of matrix
conv2d.weight = Parameter(torch.tensor([[[[1, -1],[-1, 1]]]]).float())
output = conv2d(A)
print(output)
# Assume that out and in_channels, kernal_size does not change anything 
# kernel i.e. weight = 3 - in 1D - vector of 3; 2D 3 x 3; 3D 3 x 3 x 3 - we previously assume that the parameters follow kernel size, but now it can be different
# kernal = 2 - vector of 2; 2D 2x2; 3D 2x2x2
```

```python
# Padding - add extra zeros on both ends
A_1d = torch.tensor([[[1,0,2,-1]]]).float()
conv1d = torch.nn.Conv1d(1, 1, kernel_size= 3, bias = False, padding = 2)
conv1d.weight = Parameter(torch.tensor([[[1,0,-1]]]).float())  # [[[[]]]] 4 dimensions
output = conv1d(A_1d)
print(output.tolist())
```

```python
# Stride - slide how many cells in each shift 
A_1d = torch.tensor([[[1,2,3,4,5]]]).float()
conv1d = torch.nn.Conv1d(1, 1, kernel_size= 3, bias = False, stride = 2)
conv1d.weight = Parameter(torch.tensor([[[1,0,-1]]]).float()) 
output = conv1d(A_1d)
print(output.tolist())
```

```python
# Max pooling and average Pooling
A_pool = torch.tensor([[[1,2,-1,1],
                  [0,1,-2,-1],
                  [3,0,5,0],
                  [0,1,4,-3]]]).float()
# avg_pool = torch.nn.AvgPool2d(2)
max_pool = torch.nn.MaxPool2d(2)
# out = avg_pool(A_pool)
out = max_pool(A_pool)
print(out.tolist())
```

```python
# Activation function - ReLU
import matplotlib.pyplot as plt 
x = torch.linspace(-10, 10, steps=50)  # in a line space learn the model in 100000 steps
relu = torch.nn.ReLU()
y = relu(x)
plt.title('ReLU')
plt.plot(x.tolist(), y.tolist())
plt.show()
```

```python
# Sigmoid
x = torch.linspace(-10, 10, steps=10000)  # in a line space (return values only 1 - 0) learn the model in 100000 steps
relu = torch.nn.Sigmoid()
y = relu(x)
plt.title('ReLU')
plt.plot(x.tolist(), y.tolist())
plt.show()
```
